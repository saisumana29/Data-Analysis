### Task 2C:
1. Plot the Age Intake column and observe the distribution.
2. Group certain values into categories,
- Less than 1 week, Less than 1 month, Less than 1 year
- Drop inappropriate negative valued years rows.
3. Plot the Column to again see the new distributuions.

Tip: consider 0 years as less than 1 year.

```r
library(readr)
library(dplyr)
library(ggplot2)
library(stringr)

convert_age_to_years <- function(age) {
  if (str_detect(age, "year")) {
    as.numeric(str_extract(age, "\\d+"))
  } else if (str_detect(age, "month")) {
    as.numeric(str_extract(age, "\\d+")) / 12
  } else if (str_detect(age, "week")) {
    as.numeric(str_extract(age, "\\d+")) / 52
  } else if (str_detect(age, "day")) {
    as.numeric(str_extract(age, "\\d+")) / 365
  } else {
    NA
  }
}

data$AgeInYears <- sapply(data$`Age upon Intake`, convert_age_to_years)

ggplot(data, aes(x = AgeInYears)) +
  geom_histogram(binwidth = 0.5, fill = "blue", color = "black") +
  theme_minimal() +
  labs(title = "Distribution of Age upon Intake (in Years)", x = "Age (Years)", y = "Frequency")
```

```r
# Load required libraries
library(readr)
library(dplyr)
library(ggplot2)

# Read the data
df <- read_csv("Animal_Center_Intakes.csv", show_col_types = FALSE)

# Define age_to_days function
age_to_days <- function(age) {
  age <- as.character(age)  # Convert to string to handle both int and str inputs
  if (grepl("years?$", age)) {
    return(as.integer(strsplit(age, " ")[[1]][1]) * 365)
  } else if (grepl("months?$", age)) {
    return(as.integer(strsplit(age, " ")[[1]][1]) * 30)
  } else if (grepl("weeks?$", age)) {
    return(as.integer(strsplit(age, " ")[[1]][1]) * 7)
  } else if (grepl("days?$", age)) {
    return(as.integer(strsplit(age, " ")[[1]][1]))
   # Handle unrecognized formats
  }
}

# Define categorize_age function
categorize_age <- function(age) {
  if (is.na(age)) {
    return('Unknown')  # Handle NA values
  } else if (age < 7) {
    return('Less than 1 week')
  } else if (age < 30) {
    return('Less than 1 month')
  } else if (age < 365) {
    return('Less than 1 year')
  } else {
    return('1 year or older')
  }
}

# Apply age_to_days function to 'Age upon Intake' column
df$days_of_breed <- sapply(df$`Age upon Intake`, age_to_days)

# Apply categorize_age function to 'days_of_breed' column
df$Age_Category <- sapply(df$days_of_breed, categorize_age)

# Filter out negative values in 'days_of_breed' column
df <- df[!is.na(df$days_of_breed) & df$days_of_breed >= 0, ]

# Plot the distribution of Age Category
ggplot(df, aes(x = Age_Category, fill = Age_Category)) +
  geom_bar() +
  labs(title = "Distribution of Age Category upon Intake",
       x = "Age Category upon Intake",
       y = "Frequency") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  scale_fill_brewer(palette = "Pastel1") +
  guides(fill = FALSE)
```

### Task 2F:
Display horizontal bar graph for Top 10 Locations where animals were found at night time that were sick.

Note: night time = 7pm to 12am

```r
library(dplyr)
library(ggplot2)
library(lubridate)
library(readr)

# Read the CSV file
data <- read_csv("Animal_Center_Intakes.csv")



# Parse datetime strings
data$parsed_datetime1  <- strptime(data$DateTime, format = c('%m-%d-%Y %H:%M'))
data$parsed_datetime2  <- strptime(data$DateTime, format = c('%m/%d/%Y %I:%M:%S %p'))
data$DateTime <- coalesce(data$parsed_datetime1, data$parsed_datetime2)


# Filter the dataset for animals found at night time and were sick
night_sick_animals <- data %>%
  filter(hour(DateTime) >= 19 & hour(DateTime) <= 24 , `Intake Condition` == "Sick")

# Count occurrences of each location
location_counts <- night_sick_animals %>%
  count(`Found Location`)

# Select top 10 locations with the highest counts
top_10_locations <- location_counts %>%
  top_n(10, wt = n)

# Plot horizontal bar graph
ggplot(top_10_locations, aes(x = n, y = reorder(`Found Location`, n))) +
  geom_bar(stat = "identity", fill = "skyblue") +
  labs(title = "Top 10 Locations of Sick Animals Found at Night", x = "Count", y = "Location") +
  theme_minimal()
```

### Task 2J:
Print Top 10 Breeds of animal who were less than a year old when they were taken in along with what type of animal they were.

```r
library(dplyr)

data <- read_csv("Animal_Center_Intakes.csv",show_col_types = FALSE)
# Function to convert age strings to days
convert_age_to_days <- function(age) {
  if (grepl("year", age)) {
    return(as.numeric(gsub(".*?(\\d+) year.*", "\\1", age)) * 365)
  } else if (grepl("month", age)) {
    return(as.numeric(gsub(".*?(\\d+) month.*", "\\1", age)) * 30)  # Assuming 30 days in a month
  } else if (grepl("week", age)) {
    return(as.numeric(gsub(".*?(\\d+) week.*", "\\1", age)) * 7)
  } else if (grepl("day", age)) {
    return(as.numeric(gsub(".*?(\\d+) day.*", "\\1", age)))
  } else {
    return(NA)  # Return NA for invalid age format
  }
}

# Convert age upon intake to days
data <- data %>%
  mutate(AgeInDays = sapply(`Age upon Intake`, convert_age_to_days))

# Filter the dataset for animals less than a year old upon intake
young_animals <- data %>%
  filter(AgeInDays < 365)

# Group by breed and animal type, and count occurrences
breed_counts <- young_animals %>%
  group_by(Breed, `Animal Type`) %>%
  summarise(Count = n(), .groups = 'drop')

# Select the top 10 breeds based on the count
top_10_breeds <- breed_counts %>%
  arrange(desc(Count)) %>%
  head(10)

# Print the top 10 breeds with their corresponding animal types
print(top_10_breeds)
```
