### Task G:

Understand this,

- antecedent support: If X is called antecendent, 'antecedent support' computes the proportion of transactions that contain the antecedent X.
- consequent support: If Y is called consequent, 'consequent support' computes the proportion of transactions that contain the antecedent Y.
- support: 'support' computes the proportion of transactions that contain the antecedent X and Y.
- confidence: Probability of buying Y when X is bought.
- lift: Represents how many times the probability of getting Y increases when X is received.

Define a function which takes the (sorted values from the association rules,item) and returns top recommendation list for that item.

Tip: the recommendated item are Consequents by matching the product antecedents as item to search for.

```python
def get_top_recommendations(rules_sorted_by_lift, item):

    filtered_rules = rules_sorted_by_lift[rules_sorted_by_lift['antecedents'].apply(lambda x: item in x)]

    recommendations = filtered_rules.sort_values(by='confidence', ascending=False)['consequents'].apply(list).tolist()

    recommended_items = list(set([item for sublist in recommendations for item in sublist]))

    return recommended_items
```

## Generate Association Rules

Minimum Support = 7, Minimum Confidence = 85%

### Association Rules:

- **Rule-1: `{Beer}` => `{potato chips}`**
  - Confidence = support(beer, potato chips) / support(beer) = (9/11) * 100 = 81% (not valid)

- **Rule-2: `{potato chips}` => `{beer}`**
  - Confidence = support(beer, potato chips) / support(potato chips) = (9/10) * 100 = 90% (valid)

- **Rule-3: `{wine}` => `{cheese}`**
  - Confidence = support(wine, cheese) / support(wine) = (7/8) * 100 = 87% (valid)

- **Rule-4: `{cheese}` => `{wine}`**
  - Confidence = support(wine, cheese) / support(cheese) = (7/8) * 100 = 87% (valid)

All the above association rules are strong and confidence is greater than the minimum confidence of 85%, except for Rule-1.


```python
# code for problem solution
from itertools import combinations
from collections import defaultdict

def load_data():
    data = [
        ["Beer", "Wine", "Cheese"],
        ["Beer", "Potato chips"],
        ["Eggs", "Flour", "Butter", "Cheese"],
        ["Eggs", "Flour", "Butter", "Beer", "Potato chips"],
        ["Wine", "Cheese"],
        ["Potato chips"],
        ["Eggs", "Flour", "Butter", "Wine", "Cheese"],
        ["Eggs", "Flour", "Butter", "Beer", "Potato chips"],
        ["Wine", "Beer"],
        ["Beer", "Potato chips"],
        ["Butter", "Eggs"],
        ["Beer", "Potato chips"],
        ["Flour", "Eggs"],
        ["Beer", "Potato chips"],
        ["Eggs", "Flour", "Butter", "Wine", "Cheese"],
        ["Beer", "Wine", "Potato chips", "Cheese"],
        ["Wine", "Cheese"],
        ["Beer", "Potato chips"],
        ["Wine", "Cheese"],
        ["Beer", "Potato chips"]
    ]
    return data

def generate_candidates(prev_candidates, k):
    candidates = []
    for i in range(len(prev_candidates)):
        for j in range(i + 1, len(prev_candidates)):
            if prev_candidates[i][:-1] == prev_candidates[j][:-1]:
                candidate = list(set(prev_candidates[i]) | set(prev_candidates[j]))
                if len(candidate) == k:
                    candidates.append(candidate)
    return candidates

def calculate_support(data, candidates):
    support_count = defaultdict(int)
    for transaction in data:
        for candidate in candidates:
            if set(candidate).issubset(set(transaction)):
                support_count[tuple(candidate)] += 1
    return support_count

def apriori(data, min_support, min_confidence):
    transactions = load_data()
    support_threshold = min_support * len(data)

    # Generate frequent itemsets of size 1
    frequent_itemsets = []
    candidate_itemsets = [[item] for item in set(item for transaction in data for item in transaction)]
    frequent_itemsets.append([itemset for itemset in candidate_itemsets if calculate_support(data, [itemset])[tuple(itemset)] >= support_threshold])

    k = 2
    while len(frequent_itemsets[-1]) > 0:
        candidate_itemsets = generate_candidates(frequent_itemsets[-1], k)
        frequent_itemsets.append([itemset for itemset in candidate_itemsets if calculate_support(data, [itemset])[tuple(itemset)] >= support_threshold])
        k += 1

    # Generate association rules
    association_rules = []
    for itemsets in frequent_itemsets[1:]:
        for itemset in itemsets:
            for i in range(1, len(itemset)):
                for subset in combinations(itemset, i):
                    antecedent = list(subset)
                    consequent = list(set(itemset) - set(antecedent))
                    antecedent_support = calculate_support(data, [antecedent])[tuple(antecedent)]
                    rule_support = calculate_support(data, [itemset])[tuple(itemset)]
                    confidence = rule_support / antecedent_support
                    if confidence >= min_confidence:
                        association_rules.append((antecedent, consequent, confidence))
    return association_rules

# Example usage
data = load_data()
min_support = 7 / len(data)
min_confidence = 0.85
association_rules = apriori(data, min_support, min_confidence)
for rule in association_rules:
    print(f"Rule: {rule[0]} -> {rule[1]}, Confidence: {rule[2]}")
```
